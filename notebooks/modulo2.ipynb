```python
pd.merge(sets, themes, left_on = 'theme_id' , right_on = 'id')
```

```python
sets_with_themes = pd.merge(sets, themes, left_on="theme_id", right_on="id")
sets_with_themes.head()
```

```python
sets_with_themes = pd.merge(sets, themes, on = 'theme_id')
sets_with_themes.head()
```

Que hacer con las columnas repetidas?
En este caso name_x y name_y

```python
sets_with_themes = pd.merge(sets, themes, on = 'theme_id')
sets_with_themes.head()
```

```python
sets_with_themes = pd.merge(sets, themes, on = 'theme_id', suffixes= ('_set' ,'_themes'))
sets_with_themes.head()
```

Eliminamos de themes la columna que no es relevante, tomando solamente la que nos interesa.

```python
themes.head()
```

```python
sets_with_themes = pd.merge(sets, themes.get(['theme_id','name']), on="theme_id", suffixes=("_set", "_themes"))
sets_with_themes.head()
```

### Ejercicio 2
Calcular la cantidad de existencias en inventario del sets del tema Star Wars

```python
sets.head(2)
```

```python
themes.head(2)
```

```python
inventory_sets.head()
```

```python
quantity=  inventory_sets.get(['set_num', 'quantity']).groupby('set_num', as_index = False).sum()
quantity.head()
```

```python
set_themes = pd.merge(sets, themes [['theme_id', 'name']], suffixes=('_sets', '_themes'), on = 'theme_id')
set_themes.head()
```

```python
themes_quantity = pd.merge(set_themes, quantity, on = 'set_num', validate= '1:1')
themes_quantity.head()
```

```python
star_wars_quantity = themes_quantity.query("name_themes == 'Star Wars'")
star_wars_quantity.head()
```

```python
star_wars_quantity['quantity'].sum()
```

```python
star_wars_quantity.sort_values(by = 'quantity', ascending = False)
```
