```python
def esAdulto(EDAD):
    """
    This function determines if a person is an adult based on their age.

    Parameters:
    EDAD (int): The age of the person.

    Returns:
    int: 1 if the person is an adult (age >= 18), 0 otherwise.
    """
    if EDAD >= 18:
        return 1
    else:
        return 0
    
```

```python
medals_subset.groupby("NOC").count()
```

pivot table

```python
medals.query('MEDAL == "Gold"')\
        .get(['NOC', 'MEDAL'])\
        .groupby('NOC', as_index = False)\
        .count()
```

```python
medals.query('MEDAL == "Gold"')\
    .get(["NOC", "MEDAL"])\
    .groupby("NOC", as_index=False)\
    .count()\
    .sort_values('MEDAL', ascending=False)\
    .head(10)
```

```python
medals.get(["NOC", "MEDAL","EVENT"])\
    .groupby(["NOC","MEDAL"], as_index=False)\
    .count()
```

```python
medals.get(['NOC','MEDAL','EVENT']).pivot_table(index = 'NOC', columns = 'MEDAL', aggfunc = 'count')
```

```python
medals_pivot = medals.get(['NOC','MEDAL','EVENT']).pivot_table(index = 'NOC', columns = 'MEDAL', aggfunc = 'count', fill_value = 0)
medals_pivot
```
