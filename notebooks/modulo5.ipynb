***for loops- iterrowa - apply - operaciones vectorizadas***

```python
olympics = pd.read_csv(r"..\datasets\03-analisis-exploratorio\olympics.csv")
olympics.head()
```

```python
%%timeit

lista_adultos= []

for age in olympics['Age']:

    if age >= 18:

        lista_adultos.append(1)
        
    else:

        lista_adultos.append(0)

olympics['Adult'] = lista_adultos
```

```python
olympics.head()
```

```python
olympics = pd.read_csv(r"..\datasets\03-analisis-exploratorio\olympics.csv")
```

```python
def esAdulto(age):

    if age >= 18:
        return 1
    
    else:
        return 0
    

```

```python
%%timeit
olympics['Adulto'] = olympics['Age'].apply(esAdulto)
```

```python
olympics = pd.read_csv(r"..\datasets\03-analisis-exploratorio\olympics.csv")
```

```python
import numpy as np
esAdultoVectorial = np.vectorize(esAdulto)
```

```python
%%timeit
olympics['Adult'] = esAdultoVectorial(olympics['Age'])
```

```python
olympics = pd.read_csv(r"..\datasets\03-analisis-exploratorio\olympics.csv")
olympics.head()
```

```python
import time
import dask.dataframe as dd

start_time = time.time()

# Cargar el archivo CSV
taxis = dd.read_csv(r"..\datasets\05-pandas-para-big-data\taxis.csv")

# Calcular la media
mean_fare = taxis["fare_amount"].mean().compute()

end_time = time.time()

print(f"La lectura del archivo demoró {end_time - start_time} segundos.")
print(f"Media de fare_amount: {mean_fare}")
```
