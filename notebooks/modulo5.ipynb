```python
chunks = pd.read_csv(r"..\datasets\03-analisis-exploratorio\olympics.csv", chunksize= 10000)
```

```python
from tqdm import tqdm
gold_medals = pd.DataFrame()

for chunk in tqdm(chunks, desc= "Procesando chunks"):
    gold_medals = pd.concat([gold_medals, chunk.query('Medal == "Gold"')], ignore_index=True)
```
